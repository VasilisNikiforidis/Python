```python
from IPython.display import clear_output
import random
###Displays the board
def display_board(board):
    clear_output()
    
    print('   |   |   ')
    print(' '+board[7]+' | '+board[8]+' | '+board[9]+' ')
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' '+board[4]+' | '+board[5]+' | '+board[6]+' ')
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' '+board[1]+' | '+board[2]+' | '+board[3]+' ')
    print('   |   |   ')
    
###Stores each players' marker for later use
def player_input():
    marker=''
    while not(marker=='X' or marker=='O'):
        marker=input('Player 1 choose your marker: X or O:\n').upper()
    
    if marker=='X':
        return ('X','O')
    else:
        return ('O','X')
    
###Places marker on the board
def place_marker(board,position,marker):
    
    board[position]=marker
    
###Checks if a position in available
def space_check(board,position):
    return board[position] == ' '

###Checks if the board is full (end-game)
def full_check(board):
    for i in range(1,10):
        if space_check(board,i):
            return False
    return True

###Checks if we have a win 
def win_check(board,marker):
    return (board[1]==board[2]==board[3]==marker) or (board[4]==board[5]==board[6]==marker) or\
    (board[7]==board[8]==board[9]==marker) or (board[1]==board[4]==board[7]==marker) or\
    (board[2]==board[5]==board[8]==marker) or (board[3]==board[6]==board[9]==marker) or\
    (board[1]==board[5]==board[9]==marker) or (board[3]==board[5]==board[7]==marker)
    
###Player chooses position
def player_choice(board):
    
    position = 0
    
    while position not in range(1,10) or not space_check(board,position):
        position = int(input('Choose where you want to place your marker: (1-9)\n'))
        
    return position
        
###Check who goes 1st
def starting_player():
    x=random.randint(1,2)
    return f'Player {x}'
        
###Ask if they want to play again
def replay():
    answer=input('Do you want to play again? (Y/N).\n')
    return answer.lower()=='y'
```

```python
print('Welcome to a game of Tic Tac Toe!!!')

while True:
    board=[' ']*10
    player1_marker,player2_marker=player_input()
    turn = starting_player()
    print(f'{turn} will go first!')
    start_game = input('Shall we start? (Y/N)\n').lower()
    if start_game == 'y':
        game_on=True
    else:
        game_on=False
        
        
    while game_on:
        
        if turn == 'Player 1':
            
            ###Player 1###
            
            display_board(board)
            print('Player 1 it is your turn!')
            position=player_choice(board)
            place_marker(board,position,player1_marker)
            if win_check(board,player1_marker):
                display_board(board)
                print('Player 1 WINS!')
                game_on=False
            else:
                if full_check(board):
                    display_board(board)
                    print('The game ended in a TIE!')
                    game_on=False
                else:
                    turn = 'Player 2'
                    
                    
        else:
            
            ###Player 2###
            
            display_board(board)
            print('Player 2 it is your turn!')
            position=player_choice(board)
            place_marker(board,position,player2_marker)
            if win_check(board,player2_marker):
                display_board(board)
                print('Player 2 WINS!')
                game_on=False
            else:
                if full_check(board):
                    display_board(board)
                    print('The game ended in a TIE!')
                    game_on=False
                else:
                    turn='Player 1'
        
        
        
    else:
        if replay():
            game_on=True
        else:
            break   
        
```
